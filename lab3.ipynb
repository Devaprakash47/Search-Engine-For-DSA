```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Error metrics
actual=np.array([12,4,6,3])
pred=np.array([2,8,6,3])

new_actual=np.array([7,1.2])
new_pred=np.array([4,9])

update_actual=np.append(actual,new_actual)
update_pred=np.append(pred,new_pred)

mae=np.mean(np.abs(update_actual-update_pred))
print(mae)
mse=np.mean((update_actual-update_pred)**2)
print(mse)
rmse=np.sqrt(mse)
print(rmse)
relrmse=rmse/np.mean(actual)
print(relrmse)
cv=rmse/np.mean(actual)*100
print(cv)

plt.figure(figsize=(8,5))
plt.scatter(range(len(update_actual)), update_actual, color='blue', label='Actual')
plt.scatter(range(len(update_pred)), update_pred, color='red', label='Predicted', marker='*')
plt.plot(range(len(update_actual)), update_actual, color='blue', alpha=0.3)
plt.plot(range(len(update_pred)), update_pred, color='red', alpha=0.3)
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.title('Actual vs Predicted Values')
plt.legend()
plt.show()
```
